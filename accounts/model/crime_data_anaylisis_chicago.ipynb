```python
import string
from nltk.corpus import stopwords
from collections import Counter
import pandas as pd

# Assuming description_dict is your dictionary of descriptions


def text_preprocess(mess):
    """
    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    2. Remove all stopwords
    3. Returns a list of the cleaned text
    """
    # Check characters to see if they are in punctuation
    nopunc = [char for char in mess if char not in string.punctuation]

    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)
    nopunc = nopunc.lower()

    # Now just remove any stopwords and non-alphabets
    nostop = [word for word in nopunc.split() if word.lower() not in stopwords.words('english') and word.isalpha()]

    return nostop

# Apply text_preprocess to each list of descriptions in the dictionary
preprocessed_dict = {}

for primary_type, descriptions in description_dict.items():
    preprocessed_descriptions = [text_preprocess(description) for description in descriptions]
    preprocessed_dict[primary_type] = preprocessed_descriptions

# Display common keywords and their counts for each primary type
for primary_type, preprocessed_descriptions in preprocessed_dict.items():
    # Flatten the list of preprocessed descriptions
    all_words = [word for sublist in preprocessed_descriptions for word in sublist]

    # Count occurrences of each word
    word_counts = Counter(all_words)

    # Display common keywords and their counts
    print(f"Primary Type: {primary_type}")
    for word, count in word_counts.most_common(10):  # Display top 10 words, adjust as needed
        print(f"{word}: {count}")
    print("\n")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
import joblib
from sklearn import metrics
import pandas as pd

# Assuming all columns are categorical
categorical_cols = ['Primary Type','Year'	,'Month',	'Location Description',	'Time of Day',	'Season']

# Create a column transformer
preprocessor = OneHotEncoder(sparse=False, handle_unknown='ignore')

X = cric4[['Primary Type','Year'	,'Month',	'Location Description',	'Time of Day',	'Season']]  # independent
y = cric4['Arrest']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=21)

# Fit the encoder and transform the data
X_train_encoded = preprocessor.fit_transform(X_train)
X_test_encoded = preprocessor.transform(X_test)

# Create and fit the classifier
classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=10)
classifier.fit(X_train_encoded, y_train)

# Save the fitted encoder and classifier to files
joblib.dump(preprocessor, '/content/arrest_encoder_1.joblib')
joblib.dump(classifier, '/content/arrest_classifier_1.joblib')

# Make predictions
y_pred = classifier.predict(X_test_encoded)

print("Accuracy:", metrics.accuracy_score(y_test, y_pred) * 100, "\n")

cm = confusion_matrix(y_test, y_pred)
print("\n----------Confusion Matrix------------------------------------")
print(cm)

# Classification Report
print("\n----------Classification Report------------------------------------")
print(classification_report(y_test, y_pred))

```
